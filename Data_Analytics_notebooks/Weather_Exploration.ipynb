```python
import pandas as pd 
import numpy as np
```

```python
Weather_df = pd.read_csv("2018_historic_weather_3.csv")
```

```python
Weather_df
```

```python
columns = list(Weather_df.columns.values)
```

```python
for column in columns:
    print("{} null values:".format(column),Weather_df[column].isnull().sum())
print("--------------------")
```

```python
Weather_df.dtypes
```

```python
Weather_df = Weather_df.drop(['city_name', 'feels_like', 'sea_level', 'grnd_level'
                       ,'rain_1h', 'rain_3h', 'snow_1h', 'snow_3h', 'weather_icon', 'lat', 'lon'], axis=1)
```

```python
Weather_df['date'] = pd.to_datetime(Weather_df['dt'],unit='s')
```

```python
Weather_df = Weather_df.drop(['dt_iso', 'dt', 'timezone', 'temp_min', 
                              'temp_max', 'weather_id' , 'weather_description' ], axis=1)
```

```python
Weather_df['date'] = pd.to_datetime(Weather_df['date'])
Weather_df['hour'] = Weather_df['date'].dt.hour
Weather_df['date'] = Weather_df['date'].dt.date
```

```python
Weather_df['hour'] = Weather_df['hour'].astype('int64')
Weather_df['date'] = Weather_df['date'].astype('datetime64[ns]')
```

```python
Weather_df.dtypes
```

```python
Weather_df.to_csv('Clean_Weather_2018.csv', index=False, header=True , mode='a')
```
