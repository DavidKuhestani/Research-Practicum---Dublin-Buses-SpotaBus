Due to the size of the leavetimes file, we will have to read it in chunks and examine each chunk individually. 

```python
chunk_count = 1
for chunk in pd.read_csv("rt_leavetimes_DB_2018.txt", ";", chunksize=1000000):
    print("Chunk:",chunk_count)
    print('Duplicate Rows:', chunk.duplicated().sum())
    suppressed = chunk[chunk['SUPPRESSED']==1]
    print('number of suppressed rows',suppressed.shape[0])
    columns = list(Sample_leavetimes_df.columns.values)
    for column in columns:
        print("{} null values:".format(column),chunk[column].isnull().sum())
    print("--------------------")
    chunk_count+=1
```

Based on these results:
- no duplicate rows 
- DATASOURCE will be dropped as it contains constant values and therefore provides no information to our model. 
- we can remove features: PASSENGERS, PASSENGERSIN, PASSENGERSOUT, DISTANCE, NOTE as they contain only null values.
- JUSTIFICATIONID contains primarily null vlaues, this is a fault code and therefore will not be necessary in our model so it will be dropped.
- LASTUPDATE is not relevant to our model as it does not provide any information on when the journey took place.
- There were no duplicate rows in any of the chunks.
- There were a number of rows where suppressed = 1, these will have to be examined and may be dropped. 


- After further examination, it seems that these rows are worth keeping as the journey times that can be retrieved form these rows still provide information to our models. While a large number of rows are suppressed, this indicates that the rows are partially suppressed and not fully suppressed as we can see in the trips data set that there are no trips where suppressed is 1. this data still provides data to our models and therefore will not be dropped.

# Cleaning the leavetimes dataframe, merging with Trips dataframe, splitting by route and appending it to a new CSV.



This cell takes roughly 30 mins to run. Do not run as it will overwrite the cleaned master dataframe.

```python
columns = ['DAYOFSERVICE','TRIPID','PROGRNUMBER','STOPPOINTID','PLANNEDTIME_ARR',
           'PLANNEDTIME_DEP','ACTUALTIME_ARR','ACTUALTIME_DEP','VEHICLEID']
chunk_count = 1
for chunk in pd.read_csv("rt_leavetimes_DB_2018.txt", ";", usecols=columns , chunksize=5000000):
    print("Chunk:",chunk_count)
    
    #convert datatypes
    chunk['DAYOFSERVICE'] = chunk['DAYOFSERVICE'].astype('datetime64[ns]')
    chunk['TRIPID'] = chunk['TRIPID'].astype('int')
    chunk['PROGRNUMBER'] = chunk['PROGRNUMBER'].astype('int')
    chunk['STOPPOINTID'] = chunk['STOPPOINTID'].astype('int')
    chunk['PLANNEDTIME_ARR'] = chunk['PLANNEDTIME_ARR'].astype('int')
    chunk['PLANNEDTIME_DEP'] = chunk['PLANNEDTIME_DEP'].astype('int')
    chunk['ACTUALTIME_ARR'] = chunk['ACTUALTIME_ARR'].astype('int')
    chunk['ACTUALTIME_DEP'] = chunk['ACTUALTIME_DEP'].astype('int')
    chunk['VEHICLEID'] = chunk['VEHICLEID'].astype('int')
    
    
#     #remove rows where suppressed is one 
#     chunk = chunk[chunk.suppressed != 1]
    
    #create features
    chunk['MONTH'] = chunk['DAYOFSERVICE'].dt.month
    chunk['DAY'] = chunk['DAYOFSERVICE'].dt.dayofweek
    chunk['Quarter'] = chunk['DAYOFSERVICE'].dt.quarter
    chunk['ARRIVAL_TIME_ACTUAL'] = pd.to_datetime(chunk["ACTUALTIME_ARR"], unit='s').dt.strftime("%H:%M:%S") 
    chunk['DEPARTURE_TIME_ACTUAL'] = pd.to_datetime(chunk["ACTUALTIME_DEP"], unit='s').dt.strftime("%H:%M:%S")
    chunk['DWELL_TIME'] = chunk['ACTUALTIME_DEP'] - chunk['ACTUALTIME_ARR']
    
    #merge dataframes together 
    chunk = pd.merge(chunk, Trips_df[['DAYOFSERVICE','TRIPID','LINEID','ROUTEID','DIRECTION']],
            on = ['TRIPID','DAYOFSERVICE'])    
    
    #append dataframe to new csv
    chunk.to_csv('Clean_Master_Dataset_copy.csv', index=False, header=True , mode='a') 

    print("Chunk",chunk_count, "Status: Complete")
    print("--------------------")
    chunk_count+=1
```

# Function to split up the dataframe by routes

```python
Clean_Master_Dataset.csv
```

```python
def Route_splitter(dataframe):
    routes = Trips_df.LINEID.unique()
    for route in routes:
        route_df = dataframe.loc[dataframe['LINEID']== route]
        route_df.to_csv("Cleaned_By_Route/route_{}.csv".format(route), index=False, header=None , mode='a')
    return route_df
```

```python
chunk_count = 1
for chunk in pd.read_csv("Clean_Master_Dataset.csv", chunksize=5000000, low_memory=False):
    print("Chunk:",chunk_count)
    Route_splitter(chunk)
    print("Chunk",chunk_count, "Status: Complete")
    print("--------------------")
    chunk_count+=1
```

```python
#this cell adds a header to each of the dataframe that have been split up by route.
headers = ['DAYOFSERVICE','TRIPID', 'PROGRNUMBER', 'STOPPOINTID',
         'PLANNEDTIME_ARR', 'PLANNEDTIME_DEP', 'ACTUALTIME_ARR', 'ACTUALTIME_DEP',
        'VEHICLEID', 'MONTH', 'DAY', 'Quarter', 'ARRIVAL_TIME_ACTUAL',
        'DEPARTURE_TIME_ACTUAL', 'DWELL_TIME', 'LINEID', 'ROUTEID', 'DIRECTION']

routes = Trips_df.LINEID.unique()
for route in routes:
    route_file = pd.read_csv("Cleaned_By_Route/route_{}.csv".format(route),header=None)
    route_file.to_csv("Cleaned_By_Route_Header/route_{}.csv".format(route),header=headers, index=False)
```
