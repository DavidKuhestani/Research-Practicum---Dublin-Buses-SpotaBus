# Merging of data

Merge stops with stop times on stopid to get the sequence of stops and their lat and long?

```python
full_merge = pd.merge(stop_times, stops,on = 'stop_id') 
```

```python
full_merge = pd.merge(full_merge, trips,on = 'trip_id')
```

```python
full_merge = pd.merge(full_merge, routes,on = 'route_id')
```

```python
pd.set_option('display.max_columns', None)
full_merge
```

```python
full_merge_short = full_merge[['trip_id', 'stop_id','stop_sequence',
                               'stop_headsign','stop_name','stop_lat','stop_lon','direction_id','route_short_name']]
```

```python
full_merge_short.nunique()
```

```python
full_merge_short['stop_id']
```

```python
def short_id(df):
    stops = df['stop_id']
    short_stop = int(stops[len(stops) - 5:])
    return short_stop
```

```python
full_merge_short
```

```python
pd.options.mode.chained_assignment = None  # default='warn'
full_merge_short['short_stop_id'] = full_merge_short.apply(short_id, axis=1)
```

```python
full_merge_short.loc[full_merge_short['short_stop_id']== '45A']
```

```python
in_go_ahead
```

```python
stops_dict_current = {}

for route in in_go_ahead:
    print("//---------------------------------------------------//")
    print("route:", route)  
    sorted_df = full_merge_short.loc[full_merge_short['route_short_name']== '{}'.format(route)]
    if len(sorted_df.direction_id.unique()) == 1:
        
        direction = sorted_df.direction_id.unique()[0]
        
        one_direction = sorted_df.loc[sorted_df['direction_id'] == direction] 

        sorted_trip = one_direction.groupby(["trip_id"]).apply(lambda x: x.sort_values(["stop_sequence"], ascending = True)).reset_index(drop=True)

        sequence = sorted_trip.loc[sorted_trip['trip_id'] == sorted_trip.trip_id.iloc[0]].short_stop_id.tolist() 

        #converting all characters to uppercase to match the historical data 
        route = route.upper()

        stops_dict_current['{}_{}'.format(route,direction)] = {}
        stops_dict_current['{}_{}'.format(route,direction)] = sequence 

        print("{}_{} ".format(route, direction), "Complete")

    elif len(sorted_df.direction_id.unique()) == 2:
        
        for direction in range(0,2):   
            
            one_direction = sorted_df.loc[sorted_df['direction_id'] == direction]

            sorted_trip = one_direction.groupby(["trip_id"]).apply(lambda x: x.sort_values(["stop_sequence"], ascending = True)).reset_index(drop=True)
            trip = sorted_trip.trip_id.iloc[0]
            sequence = sorted_trip.loc[sorted_trip['trip_id'] == trip].short_stop_id.tolist() 

            #converting all characters to uppercase to match the historical data 
            route = route.upper()

            stops_dict_current['{}_{}'.format(route,direction)] = {}
            stops_dict_current['{}_{}'.format(route,direction)] = sequence 

            print("{}_{} ".format(route, direction), "Complete")
```

```python
stops_dict_current
```

```python
with open('Stop_sequences_full_current_GOAHEAD_GTFS.json', 'w', encoding='utf-8') as f:
    json.dump(stops_dict_current, f, ensure_ascii=False, indent=4)
```
